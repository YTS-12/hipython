```python
# 계산기 만들기
# 숫자 + 숫자 --> 결과입니다.
num1, opr, num2 = input('수식을 입력하세요. (예. 20 + 20)').split()
num1, opr, num2
```

```python
num1 = float(num1)
num2 = float(num2)
```

```python
result = 0
if opr == '+':
    result = num1+num2
elif opr == '-':
    result = num1 = num2
elif opr == '*':
    result = num1*num2
elif opr == '/':
    result = num1/num2
else:
    print('사칙연산을 지원합니다.')
print (f'{num1} {opr} {num2}의 결과는 {result} 입니다.')
```

```python
#재고관리
inventory = {'노트북':5, '마우스':120, '키보드':0, '모니터':15}
# 0 품절, 10< 부족, <50 적정, >50 충분
for i, n in inventory.items():
    if n == 0:
        s='품절'
    elif n < 10:
        s='부족'
    elif n < 50:
        s='적정'
    else:
        s='충분'    
    print(f'{i}: {n}개, {s}')
```

```python
sentence = input('문장을 입력하세요:')
words = ['파이썬', '코딩', '데이터', 'AI']

found_word = None
for word in words:
    if word in sentence:
        found_word = word
        break    # for, while 을 중간에 강제로 종료하고 빠져나올 때 사용
if found_word:
    print(f"입력한 문장에서 '{found_word}' 단어를 찾았습니다!")
else:
    print("입력한 문장에 해당 단어가 없습니다.")
```

```python
#가위바위보 게임
player = '가위'
computer = '바위'
result = '승' if ((player == '가위' and computer == '보') or\
                 (player == '바위' and computer == '가위') or\
                 (player == '보' and computer == '바위')) else\
                 '무승부' if player == computer else \
                '패'
print(result)  #패
```

```python
import random

# 컴퓨터는 랜덤으로 선택
choices = ['가위', '바위', '보']
computer = random.choice(choices)

# 사용자는 입력(잘못 입력하면 다시 입력)
player = input("가위/바위/보 중 하나를 입력하세요: ").strip()
while player not in choices:
    print("잘못 입력했어요. '가위', '바위', '보' 중에서 다시 입력하세요.")
    player = input("가위/바위/보 중 하나를 입력하세요: ").strip()

# 승/무/패 판정(원래 코드 로직 그대로)
result = '승' if ((player == '가위' and computer == '보') or
                 (player == '바위' and computer == '가위') or
                 (player == '보' and computer == '바위')) else \
         '무승부' if player == computer else \
         '패'

print(f"player: {player}, computer: {computer}, 결과: {result}")
```

```python
numbers = [1, 3, 5, 7, 9, 11, 13, 15]
search_number = 9
idx = 1
for num in numbers:
    if num == search_number:
        print(f"{idx}번째에서 찾았습니다!")
        break  # 가장 가까운 반복문에게 영향을 미침.
    print(f"현재 검사 중인 숫자: {num}")
    idx = idx+1
```

```python
def getconnections():
  con = pymysql.connections.Connection(
    host='localhost',
    user='root',
    password='0528',
    db='wntrade',
    charset='utf8'
)
  return con
```

```python
import pandas as pd
```

```python
sql = """ SELECT 담당자명
	  ,고객회사명
      ,마일리지
      ,고객.도시
      ,도시_평균마일리지
      ,(도시_평균마일리지 - 마일리지) AS 차이
FROM 고객
	,(SELECT 도시 ,AVG(마일리지) AS 도시_평균마일리지
    FROM 고객
    GROUP BY 도시) AS 도시별요약
WHERE 고객.도시 = 도시별요약.도시""" 
```

```python
getconnections()
```

```python
# 데이터베이스 연결은 항상 try로 감싸기


con = getconnections()
try:
    with con.cursor() as cursor:
        cursor.execute(sql)
        result = cursor.fetchall()
        result_df = pd.DataFrame(result)
            
finally:
    con.close()
```

```python
while True:
    s = input("정수의 목록을 공백으로 구분해서 입력하세요. 종료:'q'").strip()
    if s.lower() == 'q':
        break                                        
    if s == "":
        print(f'입력한 숫자가 없습니다. 정수를 입력해주세요.')
        continue
   
    t = s.split()
    nums = [int(k) for k in t]   #컴프리헨션
    total = 0
    p_n=[]
    n_n=[]
    for n in nums:
        if n == 0:
            continue
        if n > 0:
            p_n.append(n)
        if n < 0:
            n_n.append(n)
        total += n
    a = len(p_n)
    b = len(n_n)
    print(f'양수: {p_n} {a}개, 음수: {n_n} {b}개 합계:{total}')

print('종료합니다.')
```

```python
# 정수 n을 입력받았을 때 n개의 피보나치 수열 출력함수 
def fibonacci_n(n):
    # n개의 피보나치 수열을 리스트로 만들어 반환
    seq = []
    a, b = 0, 1
    for _ in range(n):
        seq.append(a)
        a, b = b, a + b
    return seq

# 입력받아 출력
n = int(input("정수 n 입력: "))
print(fibonacci_n(n))
```

```python
s = input("문자열 입력: ").strip()

result = ""
count = 1

for i in range(1, len(s)):
    if s[i] == s[i-1]:
        count += 1
    else:
        result += s[i-1] + str(count)
        count = 1

# 마지막 문자 처리
if s:
    result += s[-1] + str(count)

print(result)
```

```python
def chk(s):
    result = True
    val_list = []
    for num in s:
        if num not in val_list:
            val_list.append(num)
        else: 
            result = False
    if len(val_list) != 10:
        result = False
    return result

chk('0128894568')
```

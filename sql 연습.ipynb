```python
import pymysql

# 새로운 방식
con = pymysql.connections.Connection(
    host='localhost',
    user='root',
    password='0528',
    db='wntrade',
    charset='utf8'
)
```

```python

# 커서 생성
cursor = con.cursor()

# 쿼리 실행 예시
sql_query = """select 사원번호, 이름, 부서명
from 사원
left join 부서
on 사원.부서번호 = 부서.부서번호"""
```

```python
cursor.execute(sql_query)
# 쿼리 결과 가져오기
result = cursor.fetchall()
# 결과 출력
for row in result:
 print(row)
```

```python
 cursor.close()
 con.close()
```

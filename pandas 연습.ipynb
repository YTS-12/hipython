# 인덱싱
[행,열] 인덱싱

* iloc[] 행기반 인덱싱 >>> [행,열]

* loc[] 이름기반(위치) 인덱싱 >>> [행이름,열이름]

* : >> start:end 까지 슬라이싱

* loc[행이름:행이름]

* iloc[행번호:행번호] >>> end 행번호는 -1 까지

```python
#열 선택 > 시리즈 추출
print(df['name'])
```

```python
# iloc로 위치 기반 행 선택
print(df.iloc[0])     # 첫 번째 행
print(df.iloc[1:4])   # 2~4번째 행
```

```python
# loc로 라벨 기반 행 선택
print(df.loc['A'])    # 인덱스가 'A'인 행
```

```python
# 조건을 사용한 행 선택
print(df[df['age'] > 25])  # 나이가 25보다 큰 행들
```

```python
#df1[['name', 'age']]
df.loc[ : , ['name', 'age']   ]  # loc[행, 열]
```

```python
df.iloc[0:3, 0:2] #행번호, 열, :은 start:end(-1)
```

```python
# ** 매개변수 딕셔너리
def print_info(name, age, city):
    print(f'{name} [{age}] {city}')
print_info('서울', '홍길동', 30)
```

```python
def print_info(**kwargs):   #딕셔너리 매개변수
    print(kwargs)
```

```python
print_info(city='서울', name='홍길동', age=30)
```

```python
def modify_list_and_number(lst, num):
    # 리스트 수정
    lst.append(4)
    # 숫자 수정 시도
    num += 1
    print(f"함수 내부 - 리스트: {lst}, 숫자: {num}")

# 테스트
my_list = [1, 2, 3]
my_number = 10

print(f"함수 호출 전 - 리스트: {my_list}, 숫자: {my_number}")
modify_list_and_number(my_list, my_number)
print(f"함수 호출 후 - 리스트: {my_list}, 숫자: {my_number}")
```

```python
1. 함수 도출 : CRUD
2. 함수 구현
3. 문제점 도출 - 함수 리팩토링
4. 전체흐름 구현 함수
```

```python
# 간식박스 생성 함수
def make_snackbox():
    snackbox = ['두쫀쿠']
    return snackbox
```

```python
# 간식박스 추가함수
def add_snackbox(sbox):    #간식리스트
    make_msg = input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요').split(',')
    sbox.extend(make_msg)
    print(sbox)
    return sbox
```

```python
def display_snackbox(sbox):
    print(sbox)
```

```python
# 메뉴출력
def display_menu():
    print('1. 간식추가')
    print('2. 간식박스 확인')
    print('3. 간식 제거')
    print('4. 주문 확정')
    
```

```python
def order_confirm():
    make_msg = input('주문을 확정하시겠습니까? (y/n)')
    if make_msg.lower() == 'y':
        return True
    else:
        return False
```

```python
#서비스의 흐름
snack_box = make_snackbox()
print('간식박스 생성완료', snack_box)
while True:
    display_menu() #
    menu_no = input('메뉴를 선택하세요')
    if menu_no == '1':
        add_snackbox(snack_box)  #간식리스트
    elif menu_no == '2':
        display_snackbox(snack_box)
    elif menu_no == '4':
        display_snackbox(snack_box)
        result = order_confirm()
        if result:
            break
print('주문을 감사드립니다.')
        
```

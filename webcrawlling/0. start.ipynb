html 속에 있는 요소를 꺼내는 방법
1. 요소를 직접 지정 : find("요소명", id="id명")
2. CSS의 요소 검색 방법 : select("선택자")

선택자란 css 에서 요소를 지정하는 방법입니다.
- #은 id, . 은 클래스를 나타냅니다. '#subject' '.contents'
- 자식은 > 로 지정합니다. 'span > b'
- 여러개 요소 목록은 ,로 나열합니다. "div, h1"
- select() 는 지정한 선택자의 요소를 모두 가져옵니다. [ 리스트로 ]
- select_one()은 지정한 선택자 요소 중 첫번째 요소만 반환합니다.

```python
h1 = soup.select_one("h1")
print(h1.get_text())
```

```python
soup.select_one("div#subject") #id로 꺼내기
```

```python
soup.select_one("div.contents") #class로 꺼내기
```

```python
subject = soup.select_one("#subject") #id로 꺼내기
print(subject.get_text())
```

```python
contents = soup.select_one(".contents")#class로 꺼내기
print(contents.get_text())
```

```python
b_tag = soup.select_one(".contents b")
print(b_tag.text)
```

```python
a = soup.select_one("div > b")
print(a.text)
```

```python
#div 태그의 id가 subject인 태그 안에 있는 b 태그

b_tags = soup.select("div#subject b") 
```

```python
# div 태그의 attr이 contents인 태그 안에 있는 b 태그
b_tags = soup.select("div.contents b")
```

# 태그로 추출하기 find()

```python
h1 = soup.find("h1")
print(h1.get_text())
```

```python
subject = soup.find("div", id="subject")
print(subject.get_text())
```

```python
contents = soup.find("div", class_="contents")
print(contents.get_text())
```

```python
b_tag = soup.find("b")
print(b_tag.get_text())
```

```python
third_div = soup.find_all("div")[2]
b_inside = third_div.find("b")
print(b_inside.get_text())
```

# 실습 : 네이버 환율 가져오기

```python
market_index = requests.get('https://finance.naver.com/marketindex')
market_index.text[:100]
```

```python
url = "https://finance.naver.com/marketindex/"

res = requests.get(url, headers={
    "User-Agent": "Mozilla/5.0"
})

res.encoding = "euc-kr"
soup = BeautifulSoup(res.text, "html.parser")

tag = soup.select_one("div.head_info > span.value")

print(tag.text)
```

```python
soup = BeautifulSoup(res.text, "html.parser")

value = soup.select_one("a.head.jpy_usd > div > span.value")
print(value.text)
```

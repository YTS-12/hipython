# Series

```python
import pandas as pd
```

```python
s1 = pd.Series([10, 20, 30, 40, 50])
```

```python
s1[0:2]
```

```python
col_list = ['A', 'B', 'C', 'D', 'E']
data = [10, 20, 30, 40, 50]
s2 = pd.Series(data, index=col_list)
```

```python
s2
```

```python
dict_data = {'A':10, 'B':20, 'C':30, 'D':40, 'E':50}
s3 = pd.Series(dict_data)
s3
```

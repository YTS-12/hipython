```python
import pandas as pd
```

```python
sql = """SELECT 도시
	  ,AVG(마일리지) AS 평균마일리지
FROM 고객
group by 도시
HAVING AVG(마일리지) > (SELECT AVG(마일리지) FROM 고객)"""
```

```python
getconnections()

```

```python
# 데이터베이스 연결은 항상 try로 감싸기


con = getconnections()
try:
    with con.cursor() as cursor:
        cursor.execute(sql)
        result = cursor.fetchall()
        result_df = pd.DataFrame(result)
            
finally:
    con.close()
```

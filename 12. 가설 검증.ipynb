```python
import numpy as np
from scipy import stats

# ==========================================
# 1ë‹¨ê³„: ë°ì´í„° ìˆ˜ì§‘ (ê°€ìƒ ë°ì´í„° ìƒì„±)
# ==========================================
np.random.seed(42)  # ê²°ê³¼ ì¬í˜„ì„ ìœ„í•œ ì‹œë“œ ì„¤ì •

# ìƒí™©: í†µì¦ ìˆ˜ì¹˜(0~100)ê°€ ì–¼ë§ˆë‚˜ ì¤„ì–´ë“¤ì—ˆëŠ”ê°€? (ìˆ˜ì¹˜ê°€ ë†’ì„ìˆ˜ë¡ íš¨ê³¼ ì¢‹ìŒ)
# ê¸°ì¡´ ì§„í†µì œ(Aê·¸ë£¹): í‰ê·  50ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
group_old = np.random.normal(loc=50, scale=10, size=30)

# ì‹ ê·œ ì§„í†µì œ(Bê·¸ë£¹): í‰ê·  55ë§Œí¼ ê°ì†Œ, í‘œì¤€í¸ì°¨ 10, í™˜ì 30ëª…
# (ìš°ë¦¬ëŠ” ì‹ ê·œ ì•½ì´ ë” ì¢‹ë‹¤ê³  ê°€ì •í•˜ê³  ë°ì´í„°ë¥¼ ë§Œë“¤ì—ˆìŠµë‹ˆë‹¤)
group_new = np.random.normal(loc=55, scale=10, size=30)

print(f"[ë°ì´í„° í™•ì¸]")
print(f"ê¸°ì¡´ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_old.mean():.2f}")
print(f"ì‹ ê·œ ì§„í†µì œ íš¨ê³¼ í‰ê· : {group_new.mean():.2f}")
print("-" * 30)


# ==========================================
# 2ë‹¨ê³„: ê°€ì„¤ ìˆ˜ë¦½ ë° ê²€ì • í†µê³„ëŸ‰ ì‚°ì¶œ
# ==========================================
# H0: New <= Old (íš¨ê³¼ê°€ ì—†ê±°ë‚˜ ê°™ë‹¤)
# H1: New > Old (ì‹ ê·œ ì•½ì´ ë” ìš°ìˆ˜í•˜ë‹¤) -> 'greater' ì˜µì…˜ ì‚¬ìš© (ë‹¨ì¸¡ ê²€ì •)

# ë…ë¦½ í‘œë³¸ T-ê²€ì • ìˆ˜í–‰
t_stat, p_value = stats.ttest_ind(
    group_new,       # ì‹ ê·œ ì•½ (ëŒ€ë¦½ê°€ì„¤ì˜ ì£¼ì¸ê³µ)
    group_old,       # ê¸°ì¡´ ì•½
    equal_var=False, # ë“±ë¶„ì‚° ê°€ì • ì•ˆ í•¨ (Welch's t-test, ì‹¤ë¬´ì ìœ¼ë¡œ ë” ì•ˆì „í•¨)
    alternative='greater' # ì¤‘ìš”: "ë” í¬ë‹¤(ìš°ìˆ˜í•˜ë‹¤)"ë¥¼ ê²€ì¦í•˜ë¯€ë¡œ ìš°ì¸¡ ê²€ì •
)

print(f"[ê²€ì • ê²°ê³¼]")
print(f"T-statistic (ì°¨ì´ì˜ í¬ê¸°): {t_stat:.4f}")
print(f"P-value (ìš°ì—°ì¼ í™•ë¥ ): {p_value:.4f}")
print("-" * 30)


# ==========================================
# 3ë‹¨ê³„: ìµœì¢… ê²°ì • (ë¹„ì¦ˆë‹ˆìŠ¤ ì˜ì‚¬ê²°ì •)
# ==========================================
alpha = 0.05  # ìœ ì˜ìˆ˜ì¤€ (ë¦¬ìŠ¤í¬ í—ˆìš© ë²”ìœ„ 5%)

print(f"[ìµœì¢… íŒë‹¨]")
if p_value < alpha:
    print(f"ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ê¸°ê° (P-value {p_value:.4f} < {alpha})")
    print("í•´ì„: ì‹ ê·œ ì§„í†µì œì˜ íš¨ëŠ¥ì´ í†µê³„ì ìœ¼ë¡œ ìœ ì˜ë¯¸í•˜ê²Œ ìš°ìˆ˜í•©ë‹ˆë‹¤.")
    print("ğŸ‘‰ Action: ì‹ ê·œ ì•½ ì¶œì‹œ ìŠ¹ì¸ / ë§ˆì¼€íŒ… í¬ì¸íŠ¸ë¡œ í™œìš©")
else:
    print(f"ê²°ê³¼: ê·€ë¬´ê°€ì„¤ ì±„íƒ (P-value {p_value:.4f} >= {alpha})")
    print("í•´ì„: í‰ê· ì€ ì¡°ê¸ˆ ë†’ì•„ ë³´ì¼ì§€ ëª°ë¼ë„, í†µê³„ì ìœ¼ë¡œ ìš°ìˆ˜í•˜ë‹¤ê³  ë³¼ ìˆ˜ ì—†ìŠµë‹ˆë‹¤ (ìš°ì—°ì¼ ê°€ëŠ¥ì„± ë†’ìŒ).")
    print("ğŸ‘‰ Action: ì¶œì‹œ ë³´ë¥˜ / ì¶”ê°€ ì„ìƒ ì‹¤í—˜ í•„ìš”")
```

# One-way ANOVA (ì¼ì›ë°°ì¹˜ ë¶„ì‚°ë¶„ì„)

```python
# ì„¸ ê°€ì§€ í•™ìŠµë²•(A/B/C) ì— ë”°ë¼ ì‹œí—˜ ì ìˆ˜ í‰ê· ì´ ë‹¤ë¥¸ì§€ ë¹„êµ(ì§‘ë‹¨ì€ ì„œë¡œ ë‹¤ë¥¸ ì‚¬ëŒë“¤).
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm
from statsmodels.stats.multicomp import pairwise_tukeyhsd

np.random.seed(42)

# ---- ì‹¤ìŠµ ë°ì´í„°(ëª¨ì˜) ----
n = 20
methods = (["A"] * n) + (["B"] * n) + (["C"] * n)
scores = np.concatenate([
    np.random.normal(70, 8, n),   # A
    np.random.normal(78, 8, n),   # B
    np.random.normal(85, 8, n)    # C
])

df = pd.DataFrame({"method": methods, "score": scores})

# ---- One-way ANOVA ----
model = ols("score ~ C(method)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("[One-way ANOVA ê²°ê³¼]")
print(anova_table)

# ---- (ì„ íƒ) ì‚¬í›„ê²€ì •: Tukey HSD ----
print("\n[Tukey HSD ì‚¬í›„ê²€ì •]")
tukey = pairwise_tukeyhsd(endog=df["score"], groups=df["method"], alpha=0.05)
print(tukey)
```

# Repeated Measures ANOVA (ë°˜ë³µì¸¡ì • ë¶„ì‚°ë¶„ì„)

```python
# ê°™ì€ ì°¸ê°€ìë¥¼ ëŒ€ìƒìœ¼ë¡œ ìš´ë™ í”„ë¡œê·¸ë¨ ì „/ì¤‘/í›„(3ì‹œì ) ì¸¡ì •í•œ ì²´ì§€ë°©ë¥  ë³€í™”ê°€ ìˆëŠ”ì§€ ë¶„ì„.
import numpy as np
import pandas as pd
from statsmodels.stats.anova import AnovaRM

np.random.seed(42)

# ---- ì‹¤ìŠµ ë°ì´í„°(ëª¨ì˜) ----
n_subjects = 15
subjects = np.arange(1, n_subjects + 1)

times = ["pre", "mid", "post"]
rows = []

for s in subjects:
    base = np.random.normal(25, 2)        # ê°œì¸ë³„ ì‹œì‘ ì²´ì§€ë°©ë¥ 
    # ì‹œê°„ íš¨ê³¼(ê°ì†Œ) + ì•½ê°„ì˜ ë…¸ì´ì¦ˆ
    vals = {
        "pre":  base + np.random.normal(0, 0.8),
        "mid":  base - 1.0 + np.random.normal(0, 0.8),
        "post": base - 2.2 + np.random.normal(0, 0.8),
    }
    for t in times:
        rows.append([s, t, vals[t]])

df = pd.DataFrame(rows, columns=["subject", "time", "fat"])

# ---- Repeated Measures ANOVA ----
aovrm = AnovaRM(data=df, depvar="fat", subject="subject", within=["time"]).fit()
print("[Repeated Measures ANOVA ê²°ê³¼]")
print(aovrm)
```

# Two-way ANOVA (ì´ì›ë°°ì¹˜ ë¶„ì‚°ë¶„ì„)

```python
# ë¹„ë£Œ ì¢…ë¥˜(3ìˆ˜ì¤€) Ã— ê´€ìˆ˜ ë°©ì‹(2ìˆ˜ì¤€) ì´ ìˆ˜í™•ëŸ‰ì— ë¯¸ì¹˜ëŠ” ì˜í–¥(ì£¼íš¨ê³¼ + ìƒí˜¸ì‘ìš©)ì„ ë¶„ì„. (ê° ì¡°í•©ì˜ ì°¸ê°€ì/ìƒ˜í”Œì€ ì„œë¡œ ë…ë¦½)
import numpy as np
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm

np.random.seed(42)

# -----------------------------
# 1) ì‹¤ìŠµ ë°ì´í„° ìƒì„±
# -----------------------------
fert_levels = ["F1", "F2", "F3"]
irrig_levels = ["Low", "High"]
n_cell = 12

rows = []
for f in fert_levels:
    for i in irrig_levels:
        base = 50
        fert_effect = {"F1": 0, "F2": 4, "F3": 8}[f]
        irrig_effect = {"Low": 0, "High": 6}[i]
        interaction = 5 if (f == "F3" and i == "High") else 0

        mean = base + fert_effect + irrig_effect + interaction
        values = np.random.normal(mean, 4, n_cell)

        for v in values:
            rows.append([f, i, v])

# âœ… 'yield'ëŠ” íŒŒì´ì¬ í‚¤ì›Œë“œë¼ì„œ ì»¬ëŸ¼ëª…ìœ¼ë¡œ ì“°ë©´ formula íŒŒì‹±ì—ì„œ SyntaxErrorê°€ ë‚  ìˆ˜ ìˆìŒ
df = pd.DataFrame(rows, columns=["fertilizer", "irrigation", "yld"])

# ë²”ì£¼í˜•ìœ¼ë¡œ ëª…ì‹œ(ì•ˆì „)
df["fertilizer"] = df["fertilizer"].astype("category")
df["irrigation"] = df["irrigation"].astype("category")

# -----------------------------
# 2) Two-way ANOVA (ì£¼íš¨ê³¼ + ìƒí˜¸ì‘ìš©)
# -----------------------------
# âœ… ì—¬ê¸°ë„ yld ì‚¬ìš©
model = ols("yld ~ C(fertilizer) * C(irrigation)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

print("[Two-way ANOVA ê²°ê³¼]")
print(anova_table)
```

# Two-way Repeated Measures ANOVA (ì´ì› ë°˜ë³µì¸¡ì • ë¶„ì‚°ë¶„ì„)

```python
# ê°™ì€ ì°¸ê°€ìê°€
#ì¡°ê±´ A(ì˜ˆ: ì¹´í˜ì¸/ë””ì¹´í˜ì¸ 2ìˆ˜ì¤€)
#ì‹œê°„(ì˜ˆ: ì˜¤ì „/ì˜¤í›„/ì €ë… 3ìˆ˜ì¤€)
#ë‘ ìš”ì¸ ëª¨ë‘ë¥¼ ë°˜ë³µ ì¸¡ì •í•  ë•Œ ì§‘ì¤‘ë ¥ ì ìˆ˜ ì°¨ì´ë¥¼ ë¶„ì„(ì£¼íš¨ê³¼ + ìƒí˜¸ì‘ìš©).

import numpy as np
import pandas as pd
from statsmodels.stats.anova import AnovaRM

np.random.seed(42)

# ---- ì‹¤ìŠµ ë°ì´í„°(ëª¨ì˜) ----
n_subjects = 12
subjects = np.arange(1, n_subjects + 1)

drink_levels = ["Decaf", "Caffeine"]     # ìš”ì¸1(ë°˜ë³µì¸¡ì •)
time_levels = ["Morning", "Afternoon", "Evening"]  # ìš”ì¸2(ë°˜ë³µì¸¡ì •)

rows = []
for s in subjects:
    base = np.random.normal(60, 5)  # ê°œì¸ë³„ ê¸°ë³¸ ì§‘ì¤‘ë ¥

    for d in drink_levels:
        drink_effect = {"Decaf": 0, "Caffeine": 6}[d]

        for t in time_levels:
            time_effect = {"Morning": 3, "Afternoon": 0, "Evening": -4}[t]

            # ìƒí˜¸ì‘ìš©(ì˜ˆ: ì¹´í˜ì¸ì€ ì €ë…ì— ë” íš¨ê³¼ê°€ í¬ë‹¤ ê°™ì€ ìƒí™©)
            interaction = 0
            if d == "Caffeine" and t == "Evening":
                interaction = 4

            score = base + drink_effect + time_effect + interaction + np.random.normal(0, 2)
            rows.append([s, d, t, score])

df = pd.DataFrame(rows, columns=["subject", "drink", "time", "score"])

# ---- Two-way Repeated Measures ANOVA ----
aovrm2 = AnovaRM(data=df, depvar="score", subject="subject", within=["drink", "time"]).fit()
print("[Two-way Repeated Measures ANOVA ê²°ê³¼]")
print(aovrm2)
```

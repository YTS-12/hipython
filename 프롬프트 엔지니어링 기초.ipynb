```python
def msg(role, text):
    msg_src = {"role" : role, "content" : [{"type":"input_text", "text":text}]}
    return msg_src
```

```python
res = client.responses.create(
    model = 'gpt-4o-mini',
    input = [
        msg("system", "너는 마케팅 전문가야"),
        msg("user", "광고문구 만들어줘")
    ],
    temperature = 0
)
res.output_text
```

* system

* assistant : 그전에 기억하고 싶은 대화를 저장 -> 래그로 , 사람이 검수, 중간산출물

```python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        {
            "role": "assistant", #헬퍼는 그대로 두고, assistant만 명시적으로 처리
            "content": [
                {
                    "type": "output_text",
                    "text": "바로 예상을 뚫고 우승한 브라질입니다."
                }
            ]
        },
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)

response.output_text
```

```python
def msg_assist(role, text) :
    msg_src = {"role": role, "content": [{"type":"output_text", "text":text}] }
    return msg_src
```

```python
response = client.responses.create(
    model="gpt-4o-mini",
    input=[
        msg("user","2002년 월드컵에서 가장 화제가 되었던 나라는 어디야?"),
        msg_assist("assistant","바로 예상을 뚫고 우승한 브라질입니다." ),
        msg("user","그 나라가 화제가 되었던 이유를 자세하게 설명해줘")
    ]
)
response.output_text
```

# 좋은 프롬프트 조건 : 역할, 목표, 조건, 출력형식

```python
sys_prompt = """
역할 : 너는 대한민국 가수 이무진이야
목표 : 20대 30대를 위한 노래 가사를 작사하는거야. 
조건 : 표절 금지. 이무진 노래가사 스타일 참고.
출력형식 : 10줄
"""
```

```python
prompt = """주어진 고장내용 텍스트로부터 고장부품, 불량유형, 조치내용을 추출하시오.
결과는 반드시 파이썬 Dictionary 형태로만 반환하시오.

ex)
고장내용: '출하장 입구 에어 50A 유니온 에어리크 재조임'
출력: {"고장부품": "50A 유니온", "불량유형": "에어리크", "조치내용": "재조임"}

고장내용: 'SHIELD ROOM 화풍기 철거부위 마감,판넬벽체문타공'
출력: {"고장부품": "SHIELD ROOM 화풍기", "불량유형": "철거부위 마감", "조치내용": "판넬벽체문타공"}
"""
```

```python
def return_label(text):
    res = client.responses.create(
        model="gpt-4o-mini",
        input=[
            msg("system", prompt),
            msg("user", "고장내용: " + text + "\n출력:")
        ],
        temperature=0
    )
    return res.output_text
```

```python
return_label(sub_df['고장내용'].loc[0])
```

```python
result = []
for con in sub_df['고장내용']:
    result.append(return_label(con))

result
```

```python
import random

# "이 동전은 앞면이 나올 확률이 50%야"라고 미리 정함 (Design)
expected_prob = 0.5 
trial_count = 10  # 10번 던지기
heads = 0   # 앞면이 나오는 건수

print(f"--- 실험 시작 (예상 확률: {expected_prob * 100}%) ---")

for i in range(trial_count):
    # 실제 동전 던지기 (0~1 사이 랜덤 숫자 뽑기)
    if random.random() < expected_prob:
        heads += 1
        result = "앞면"
    else:
        result = "뒷면"
    print(f"{i+1}번째 결과: {result}")

# --- 3. 상대도수 (Relative Frequency): 실제 결과 ---
actual_freq = heads / trial_count

print("-" * 30)
print(f'앞면이 나온 횟수 : {heads}')
print(f"우리의 예상 (확률): {expected_prob * 100}%")
print(f"실제 결과 (상대도수): {actual_freq * 100}%")
```

```python
import math

# 1. 설정값 정의
total_balls = 7
white_balls = 4
black_balls = 3
pick_count = 3

# 2. 전체 경우의 수: C(7, 3)
total_cases = math.comb(total_balls, pick_count)

# 3. 사건의 경우의 수: C(4, 2) x C(3, 1)
# 흰공 2개 뽑기 x 검은공 1개 뽑기
event_cases = math.comb(white_balls, 2) * math.comb(black_balls, 1)

# 4. 확률 계산
probability = event_cases / total_cases

print(f"계산된 확률 (분수): {event_cases}/{total_cases}")
print(f"계산된 확률 (소수): {probability:.2f}")
```

```python
import random
def run_simulation(trials):
    urn=['W', 'W', 'W', 'W', 'B', 'B', 'B']
    success_count = 0
    for _ in range(trials):
        drawn_balls = random.sample(urn,3)
        if drawn_balls.count('W') == 2 and drawn_balls.count('B') == 1:
            success_count += 1
    return success_count / trials

trials=int(input(f'시뮬레이션 횟수를 입력하세요.'))
simulated_prob = run_simulation(trials)

print(f"시뮬레이션 횟수: {trials}번")
print(f"실험 결과 확률: {simulated_prob:.5f}")
print(f"이론적 확률(0.5143)과의 오차: {abs(simulated_prob - 18/35):.5f}")
```

# 확률의 곱셈법칙

```python
# --- Case 1: 독립사건 (주사위와 동전) ---
# 서로 영향을 주지 않으므로 각각 정의하고 그냥 곱합니다.
prob_dice_even = 0.5  # 주사위 짝수
prob_coin_head = 0.5  # 동전 앞면

# P(A ∩ B) = P(A) * P(B)
prob_independent = prob_dice_even * prob_coin_head
print(f"독립사건 확률: {prob_independent}")
```

```python
# --- Case 2: 종속사건 (제비뽑기: 당첨 2, 꽝 8) ---
# 앞사람이 뽑은 결과가 뒷사람 확률을 바꿉니다.
total = 10
win = 2

# A: 첫 번째 사람이 당첨될 확률 P(A)
prob_A = win / total  # 2/10

# B|A: (A가 당첨된 후) 두 번째 사람도 당첨될 확률 P(B|A)
# 전체도 1 줄고, 당첨제비도 1 줄어듭니다. (상태 변경!)
prob_B_given_A = (win - 1) / (total - 1) # 1/9

# P(A ∩ B) = P(A) * P(B|A)
prob_dependent = prob_A * prob_B_given_A
print(f"종속사건 확률: {prob_dependent:.3f}")
```

# 조건부 확률

# 기대값 구하기

```python
x_values = [1,2,3]
p_x = [1/6, 3/6, 2/6]

exp_value = 0
#기대값
for i in range( len(x_values)):
  exp_value += x_values[i] * p_x[i]

print(f'기대값 : {exp_value:.5f}')
```

```python
# 카드 점수
cards = {"A": 10, "B": 20, "C": 50}

# 모두 같은 확률(무작위로 1장)
p = 1 / len(cards)

# 기대값 E[X] = Σ (값 * 확률)
expected_value = sum(score * p for score in cards.values())

print(f"기대값 E[X] = {expected_value}")
```

# 리스트와 for문의 활용

```python
students = [{'num':'1','name':'kim', 'kor':80, 'math':90, 'eng':100,},
            {'num':'2','name':'park', 'kor':90, 'math':60, 'eng':80,},
            {'num':'3','name':'hong', 'kor':70, 'math':70, 'eng':80,},
            {'num':'4','name':'hong1', 'kor':70, 'math':70, 'eng':80,},
            {'num':'5','name':'hong2', 'kor':70, 'math':70, 'eng':80,}]        # {}  딕셔너리
```

```python
#인덱싱하기
students[2]['kor']
```

```python
print(students)
```

```python
for s in students : 
    print(s)
    print('\n')
```

```python
for s in students:
    s['total'] = s['kor'] + s['math'] + s['eng']
    print(s['total'])
    print('\n')
```

```python
for s in students : 
    print(s)
    print('\n')
```

```python
for s in students:
    s['mean'] = round((s['total'])/3,2)
    print(s['mean'])
    print('\n')
```

```python
for s in students:
    print(s)
```

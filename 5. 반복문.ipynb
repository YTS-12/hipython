```python
#ë‚ ì§œ ì…ë ¥ë°›ê¸° > ìš”ì¼ ì¶œë ¥í•˜ê¸°
year = int(input('ë…„ë„ë¥¼ ì…ë ¥í•˜ì‹œì˜¤:'))
month = int(input('ì›”ì„ ì…ë ¥í•˜ì‹œì˜¤:'))
day = int(input('ì¼ì„ ì…ë ¥í•˜ì‹œì˜¤:'))

total_days = 0
for i in range(1,year):
    if i % 400 == 0: 
        total_days += 366
    elif i % 100 == 0:
        total_days += 365
    elif i % 4 == 0:
        total_days += 366
    else:
        total_days += 365
print(f'1st : {total_days}, {year}')

month_days = [0,31,28,31,30,31,30,31,31,30,31,30,31]
for m in range(1,month):
    total_days += month_days[m]
print(f'2nd : {total_days}, {year}')
if month >= 3 : 
    if year % 400 == 0: 
        total_days += 1
    elif year % 100 == 0:
        total_days += 0
    elif year % 4 == 0:
        total_days += 1
    else:
        total_days += 0
print(f'3nd : {total_days}, {year}, {month}')

total_days += (day-1)

print(f'final : {total_days}, {year}, {month}, {day}')

remain = (total_days % 7)
d = ['ì›”ìš”ì¼', 'í™”ìš”ì¼', 'ìˆ˜ìš”ì¼', 'ëª©ìš”ì¼', 'ê¸ˆìš”ì¼', 'í† ìš”ì¼', 'ì¼ìš”ì¼']
result = d[remain]
print(f'{year}ë…„ {month}ì›” {day}ì¼ì€ {result}ì…ë‹ˆë‹¤.')
```

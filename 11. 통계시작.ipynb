## 평균

```python
# 산술평균 구하기 
import numpy as np
data = [10,20,30,40,50]
# sum(data)/len(data)
np.mean(data)
```

```python
# 가중평균 구하기
data_2 = {'점수': [97, 94, 92, 91, 90, 89, 78, 60],'인원수': [4, 11, 12, 21, 30, 12, 9, 1]}
weighted_mean = np.average(data_2['점수'], weights=data_2['인원수'])
print(weighted_mean)
```

```python
# 기하평균 구하기
import numpy as np

# 초기 매출
initial = 100

# 시나리오 적용
step1 = initial * 2.0      # +100%
step2 = step1 * 0.5        # -50%
final = step2

print(f"초기: {initial}만원")
print(f"1차 후: {step1}만원")
print(f"최종: {final}만원")
print(f"\n실제 변화: {((final/initial - 1) * 100):.1f}%")

# 산술평균
arithmetic = (100 + (-50)) / 2
print(f"산술평균: {arithmetic:.1f}%")

# 기하평균
geometric = (np.sqrt(2.0 * 0.5) - 1) * 100
print(f"기하평균: {geometric:.1f}%")
```

```python
# 조화평균 구하기
import numpy as np
from scipy import stats

# 1. 데이터: 갈 때 100km/h, 올 때 50km/h
speeds = [100, 50]

# 2. 평균 계산 비교
# A. 산술평균 (오답)
arithmetic = np.mean(speeds)

# B. 조화평균 (정답)
harmonic = stats.hmean(speeds)

# 3. 결과 리포트
print(f"데이터: {speeds}")
print("-" * 35)
print(f"1. [❌ 오류] 산술평균: {arithmetic:.2f} km/h")
print("-" * 35)
print(f"2. [✅ 정답] 조화평균: {harmonic:.2f} km/h")
print(f"   >> 해석: '느린 구간(50) 때문에 전체 효율이 급감했습니다.'")
print(f"   >> 실제 물리적 평균 속도는 66.67km/h가 맞습니다.")
```

```python
# 중앙값 구하기
import numpy as np

data = [10, 20, 30, 40, 50]
median_value = np.median(data)
print(median_value)  # 30
```

```python
# 중앙값과 백분위
import numpy as np

scores = [65, 75, 80, 85, 95]

# 백분위수 계산
p25 = np.percentile(scores, 25)  # Q1
p50 = np.percentile(scores, 50)  # 중앙값
p75 = np.percentile(scores, 75)  # Q3

print(f"25번째 백분위수: {p25}")  # 75.0
print(f"50번째 백분위수(중앙값): {p50}")  # 80.0
print(f"75번째 백분위수: {p75}")  # 85.0
print(f"사분위간 범위(IQR): {p75 - p25}")  # 10.0
```

```python
# 백분위수 (Percentile) 
# 백분위수는 데이터를 크기순으로 나열하여 100등분 했을 때, 특정 백분율에 해당하는 위치의 값을 의미합니다. 
# 의미: '제 n 백분위수'는 전체 데이터 중 n%가 해당 값보다 작거나 같음을 나타냅니다.\
# 예시:제10 백분위수: 하위 10%에 해당하는 지점의 값입니다.
# 제50 백분위수: 중앙값과 동일합니다.
# 제90 백분위수: 상위 10% 수준(나보다 작은 값이 90%)을 의미합니다.
```

```python
# 최빈값 구하기 
scores = [70, 80, 80, 90, 60, 80, 70]
counts = {}

for x in scores:
    if x in counts:
        counts[x] += 1
    else:
        counts[x] = 1

counts
# 결과
{70: 2, 80: 3, 90: 1, 60: 1}
#최빈값은 ? 80

#함수로 구하기
mode = max(counts, key=counts.get)
mode
```

## 첨도
* 양의 첨도 : 정규분포보다 값의 크기만큼 뾰족하다. 꼬리가 두껍다(길다)
* 음의 첨도 : 정규분포보다 값의 크기만큼 눌려있다. 꼬리가 얇고 짧다. 평균주변에 고르게 있다.

```python
from scipy.stats import kurtosis
positive_kurtosis = [
    50, 50, 50, 50, 50,
    48, 52,
    35, 65
]
negative_kurtosis = [
    42, 44, 46, 48,
    52, 54, 56, 58
]
pos_kurtosis_value = kurtosis(positive_kurtosis )
neg_kurtosis_value = kurtosis(negative_kurtosis )

print(pos_kurtosis_value, neg_kurtosis_value )
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"

column_names = [
    'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX',
    'PTRATIO','B','LSTAT','MEDV'
]

df = pd.read_csv(url, sep=r'\s+', header=None, names=column_names)

# 1) 상관계수(피어슨) 계산
corr = df['PTRATIO'].corr(df['MEDV'])
print(f"PTRATIO와 MEDV의 상관계수(Pearson): {corr:.4f}")

# ✅ 추세선(1차 회귀선) 계산
x = df['PTRATIO'].values
y = df['MEDV'].values
m, b = np.polyfit(x, y, 1)  # y = m*x + b
x_line = np.linspace(x.min(), x.max(), 200)
y_line = m * x_line + b

# 2) 산점도 + 빨간색 추세선
plt.figure(figsize=(10, 6))
plt.scatter(df['PTRATIO'], df['MEDV'], label='Data')
plt.plot(x_line, y_line, color='red', linewidth=2, label='Trend line')  # ✅ 빨간 추세선
plt.title(f"PTRATIO vs MEDV (corr = {corr:.4f})")
plt.xlabel("PTRATIO (Pupil-Teacher Ratio)")
plt.ylabel("MEDV (Median value of owner-occupied homes)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
column_names = [
    'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX',
    'PTRATIO','B','LSTAT','MEDV'
]
df = pd.read_csv(url, sep=r'\s+', header=None, names=column_names)

# 1) 상관계수(피어슨) 계산
corr = df['PTRATIO'].corr(df['MEDV'])
print(f"PTRATIO와 MEDV의 상관계수(Pearson): {corr:.4f}")

# ✅ 추세선(1차 회귀선) 계산
x = df['PTRATIO'].values
y = df['MEDV'].values
m, b = np.polyfit(x, y, 1)
x_line = np.linspace(x.min(), x.max(), 200)
y_line = m * x_line + b

# ✅ 상관관계(|r|)를 반영해서 선의 진하기/두께 조절
strength = abs(corr)                 # 0~1
alpha_val = 0.2 + 0.8 * strength     # 최소 0.2 ~ 최대 1.0
lw = 1 + 4 * strength                # 최소 1 ~ 최대 5

# 2) 산점도 + 상관관계 반영된 빨간 점선 추세선
plt.figure(figsize=(10, 6))
plt.scatter(df['PTRATIO'], df['MEDV'], label='Data')
plt.plot(x_line, y_line, color='red', linestyle='--',
         linewidth=lw, alpha=alpha_val,
         label=f"Trend (|r|={strength:.2f})")

plt.title(f"PTRATIO vs MEDV (corr = {corr:.4f})")
plt.xlabel("PTRATIO (Pupil-Teacher Ratio)")
plt.ylabel("MEDV (Median value of owner-occupied homes)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"

column_names = [
    'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX',
    'PTRATIO','B','LSTAT','MEDV'
]

df = pd.read_csv(url, sep=r'\s+', header=None, names=column_names)

# 1) 상관계수(피어슨) 계산
corr = df['PTRATIO'].corr(df['MEDV'])
print(f"PTRATIO와 MEDV의 상관계수(Pearson): {corr:.4f}")

x = df['PTRATIO'].values
y = df['MEDV'].values

x_line = np.linspace(x.min(), x.max(), 200)

# ✅ 1차 추세선
m1, b1 = np.polyfit(x, y, 1)
y_line1 = m1 * x_line + b1

# ✅ 2차 추세선(추가)
a2, b2, c2 = np.polyfit(x, y, 2)          # y = a2*x^2 + b2*x + c2
y_line2 = a2 * x_line**2 + b2 * x_line + c2

# 2) 산점도 + 1차/2차 추세선
plt.figure(figsize=(10, 6))
plt.scatter(df['PTRATIO'], df['MEDV'], label='Data')

plt.plot(x_line, y_line1, color='red', linewidth=2, label='Linear trend (1st)')
plt.plot(x_line, y_line2, color='red', linestyle='--', linewidth=2, label='Poly trend (2nd)')  # ✅ 2차

plt.title(f"PTRATIO vs MEDV (corr = {corr:.4f})")
plt.xlabel("PTRATIO (Pupil-Teacher Ratio)")
plt.ylabel("MEDV (Median value of owner-occupied homes)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"

column_names = [
    'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX',
    'PTRATIO','B','LSTAT','MEDV'
]

df = pd.read_csv(url, sep=r'\s+', header=None, names=column_names)

# 1) 상관계수(피어슨) 계산
corr = df['PTRATIO'].corr(df['MEDV'])
print(f"PTRATIO와 MEDV의 상관계수(Pearson): {corr:.4f}")

x = df['PTRATIO'].values
y = df['MEDV'].values

x_line = np.linspace(x.min(), x.max(), 200)



# ✅ 3차(곡선) 추세선 추가
a3, b3, c3, d3 = np.polyfit(x, y, 3)              # y = a3*x^3 + b3*x^2 + c3*x + d3
y_line3 = a3 * x_line**3 + b3 * x_line**2 + c3 * x_line + d3


plt.figure(figsize=(10, 6))
plt.scatter(df['PTRATIO'], df['MEDV'], label='Data')


plt.plot(x_line, y_line3, color='red', linestyle=':', linewidth=2, label='Poly trend (3rd)')  # ✅ 3차

plt.title(f"PTRATIO vs MEDV (corr = {corr:.4f})")
plt.xlabel("PTRATIO (Pupil-Teacher Ratio)")
plt.ylabel("MEDV (Median value of owner-occupied homes)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
